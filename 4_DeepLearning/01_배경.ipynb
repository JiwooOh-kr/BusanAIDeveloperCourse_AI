## History

- 특징(발전 -> 암흑기 -> 발전 -> 암흑기 -> 발전 -> ...)
- year
  - 1956 : 첫 등장, 존 매카시 주장
  - 1980 : 컴퓨터 발전(IBM, 매킨토시(맥))
  - 1990 : 역전파 알고리즘 등장(y에서 x로 거슬러 가면서 식이 성립될 수 있게 가중치를 조정하는 방식 -> 최적화)
  - 1997 : IBM 딥블루라는 AI가 체스 세계 챔피언을 이겼다. AI가 인간을 이긴 첫번째 사례
  - 2006 : 합성곱 인공 신경망이 논문으로 처음 등장 -> 이때부터 딥러닝이라고 불렀다.
  - 2011 : 퀴즈대회에서 [IBM Watson](https://https://ko.wikipedia.org/wiki/%EC%99%93%EC%8A%A8_(%EC%BB%B4%ED%93%A8%ED%84%B0))이 우승 => 의료, 금융 쪽으로 Watson 진출
  - 2012 : 이미지 인식 대회(사진 속에 한 개의 이미지를 사람보다 더 잘 인식하는 대회, 카테고리가 약 1000개), 2012까지 주류는 머신러닝이었고, 이때 딥러닝으로 처음 참가한 신경망이 존재, 이후부터는 주류가 머신러닝에서 딥러닝으로 바뀌었다. (참고 : 2014년도 구글 인셉션 V2가 우승) 2017년도 종료
  - 2016 : 알파고 vs 이세돌 -> 강화학습법 흥행
  - 현재 : 전통적인 신경망(CNN, RNN, ...), 강화학습(DQN, DBQN, ...), 생성모델(GAN, ...)로 카테고리가 나눠지고 있다.

# 엔진 방식


- Define-and-Run
  - Tensorflow
  - 계산 그래프, 데이터 플로우 그래프 설계
  - 세션을 오픈하고 그래프에 데이터를 주입
  - 세션 실행 -> C++ 작동(연산은 C++이 수행)
  - 파이썬은 전체적인 틀, 세션 가동, 주입, 행위만 지시, 결과를 받고, 외부적인 작어만 수행 / 실제적인 일은 C++이 다 한다.
  - 구조에 대해서 정의를 하고, 그리고 나중에 실행한다. -> 뭔가 잘못되었는지는 실행해 봐야 안다. -> 디버깅이 쉽지 않다.
  - 텐서플로에서도 Define-by-Run 방식을 지원하게 해준다.

- Define-by-Run
  - PyTorch
  - Chainer 엔진으로부터 영향을 받았다. 일본에서 많이 사용하는 방식이다.
  - 계산 그래프를 만들면서 동시에 주입할 데이터를 넣어서 처리
  - 연산 실행을 즉시 수행 => 파이썬 짜는 방식과 동일
  - 구현이 단순화되고, 디버깅 시 오류를 잡을 수 있다(오류 처리가 쉽다).
  - 유연성, 직관적, 고성능(공통)

### RNN

- 순환(재귀적) 뉴럴 네트워크
- 기존 신경망 + 시간 개념(정보, 시퀀스)
- 자연어 처리
  - 문서 분류, 대화문 생성, 기계식 번역, 문서 요약, 챗봇
  - 자연어 처리 프로세스
    - 형태소 분석기 생성, 사전 생성, 형태소 분리(분절), 데이터의 벡터화(원-핫 인코딩) -> 수치화 -> 신경망 구성 -> 학습
  - 동영상 분류
  - [음성 인식](http://tcpschool.com/deeplearning/deep_sound)
    - AI 스피커
      - 알렉사 : 아마존, 1위 제품
  - [음성 합성](https://github.com/tensorflow/magenta), [음악 생성](https://deepmind.com/blog/article/wavenet-generative-model-raw-audio)
  - 시계열
    - 연속적인 데이터
      - 금융(주가, 환율, ...), 제조(설비, 센서, 공정, ...), 기타 모든 장비에서 나오는 데이터
    - RNN의 약점을 개선한 LSTM이 주로 많이 사용된다.
      - LSTM 외에 몇가지 더 신경망이 존재한다.
  - 대표 결과물
    - one - to - many
      - 하나를 입력해서 여러 개의 출력이 나온다.
      - 이미지 캡션, 이미지를 보고 텍스트를 생성
    - many - to - one
      - 여러 개를 입력해서 하나를 출력
      - 여러 개의 리뷰를 넣어서 긍정/부정 판별 => 감성분석(분류)
      - 스팸메일 분류
    - many - to - many
      - 여러 개를 입력해서 여러 개를 출력
      - 입력 문장에 대해서 대답 문장을 만든다. => 챗봇
      - 입력 문장으로부터 번역된 문장을 만든다. => 기계 번역

# 퍼셉트론

- 개요
  - 1957년 고안
  - 기계학습의 기초 개념
- 단순 퍼셉트론
  - 입력층과 출력층으로 구성
  - x0, x1, x2, ...의 입력을 가지고 A라는 출력으로 수렴한다(출력한다).
  - 어떤 입력 기반이 어떤 출력으로 확정되는가?

- 예시
  - 2020년 11~12월 iphone12가 출시된다. 카메라 4개에 라이더 1개를 탑재하는 등 AR 기능, 5G가 탑재되어 최고 스펙으로 강화된 폰을 출시할 것이다. 당신은 구매를 할 것인가?
  - 출력(y) : 구매한다(1), 구매하지 않는다(0)
  - 입력요인 : 구매해도 좋을지 관련된 판단 요인
    - x0 : 돈이 충분히 있는가?
    - x1 : 저장공간이 충분한가?
    - x2 : 카메라 기능이 월등한가?
    - x3 : 무게는 가벼운가?
    - x4 : 지금 사용하는 폰이 문제가 있는가? ...
  - 일반적인 다수결의 원칙을 적용한다면, 과반이 되서 구매한다면 합당한가?(돈이 충분하지 않는 경우라면 곤란해짐) => 과반 적용은 x0 ~ xn 중요도(가중치)가 동일해야 가능하다. 그러나 개별 입력 요소의 가중치는 다르다! -> 가중치를 W라고 하면
  - 구매요인 = x0W0 + x1W1 + ... + xnWn
  - 구매를 결정하는 역치(임계치, threshold, 선택의 기준값)를 C라고 한다면
  - 식
    - if 구매요인 > C : 구매
  - 이런 구매 요인을 여러 층으로 구성하여 조합하면 좋은 판단을 할 수 있다.

- Sigmoid는 제외하고 확인
- y = xW + b
- 딥러닝을 통해서(신경망이 학습을 통해서) 최적의 예측을 할 수 있는 W/b를 찾아낸다.
- 전이학습이란 이미 잘 만들어진 모델을 가져와서, W를 초기화하고 새로운 데이터를 넣어서 다시 W를 찾아가는 과정 + Customize → 이미 잘 만들어진 모델을 활용하는 전략

- 인공신경망을 이용하여 학습 후 예측하는 방식이 딥러닝
- 인공신경망은 인간의 뉴런을 본따서 만들었다.
- 활성화함수(Sigmoid)
  - 아이디어 : 신경을 타고 데이터가 이동할 때 뉴런이 활성화되는 형상
  - 데이터를 비선형으로 만들어서 예측이 쉽게 안 되게 하여 깊이를 더 주게 되는 기법
  - 레이어가 더 깊어진다.




- 인공신경망
  - 입력층 + 은닉층 + 출력층 => ANN
  - 입력층 + 은닉층1 + 은닉층2 + 출력층 => DNN(Deep)
- 순전파
