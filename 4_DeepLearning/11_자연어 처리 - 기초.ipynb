# 딥러닝 발전 분야


- 이미지 분류
  - CNN 기반 알고리즘
    - 한 장의 이미지상에서 한 개의 사물을 인식(분류)
      - 기본 CNN
      - resNet, VGG-16 등
    - 한 장의 이미지상에서 여러 사물을 분류
      - Yolo, fastCNN 등
- 기계 번역
  - 자연어 처리에서 가장 궁극의 목적지
  - 딥러닝에 의해서 어느 정도 완성이 되었다.(but 계속 진행중)
  - 기술의 발전 history
    - PBMT(규칙 기반 기계 번역 : 1950~)
    - SMT(통계 기반 기계 번역, 단어/구문 기반 : 1990~)
    - 신경망 기반(2015~)
      - RNN
- 음성 인식
  - 딥러닝 기술을 이용하여서 상용화까지 진행
  - 기술의 발전 history
    - GMM(음소인식) + HMM(시퀀스 설정 알고리즘) + AM(HMM을 이용한 데이터를 학습하여서 구성 : 음향모델) + LM(n-gram 기반 언어모델) + WFST(자동 음성 인식 ASR 시스템) = 결합되어서 생성(복잡, 난이도 높음)
    - 현재 : DNN(심층 신경망)으로 대체 -> 음향 모델은 대체적으로 LSTM(RNN의 문제점을 개선한 모델)을 사용
- 생성 모델
  - 현재 유행하는 모델
  - 데이터 X에 대해서 Y를 찾는(분류, 예측) 것이 아니라, X의 분포에 집중하여 X 자체를 묘사할 수 있는 부분에 집중하는 방식
    - 적대적 학습(GAN), ...
    - 변분 오토 인코더(VAE), ...
    - 영상을 만들 것인가, 음성을 만들 것인가, 텍스트를 만들 것인가, 이미지를 만들 것인가 등 생성의 주체에 포인트

# 자연어 처리 패러다임

- 딥러닝
  - 자연어 처리에 주류
  - 단어 임베딩 등을 통해 <u>연속적인</u> 벡터로 표현이 가능하다.
  - 자연어는 텍스트다. 딥러닝의 데이터는 무조건 수치여야 한다. => 자연어를 수치로 표현해야 학습이 가능하다.
    - 말뭉치 -> 분절 -> 사전 생성 -> 인덱스를 이용해서 수치로 표현 -> 정규화 -> 학습 데이터로써 준비 완료
    - 형태소 분석기 : NLTK(영어권), [KoNLPy](https://konlpy-ko.readthedocs.io/ko/v0.4.3/)(한국어 : Hnnanum. Kkma, Komoran, <u>**Mecab**</u>, Twitter)
    - 한국어의 표현의 어려움(모호성, 유의성)에 대한 해결이 가능해졌다.
    - RNN의 단점을 보완한 LSTM, GRU를 활용해 더 고도화 수행을 할 수 있다.(기타 기법(어텐션 등)을 혼용하여 더욱더 발전했다.)
