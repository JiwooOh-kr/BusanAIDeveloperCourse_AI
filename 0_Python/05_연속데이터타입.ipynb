# 리스트
- [], 순서가 존재(0,1,2,~ -1,-2,...), 값의 중복 ok, 수정 ok, 삭제 ok, 추가 ok


```
# 생성
nums = [] # 정적인 형태로 생성 : 리스트 생성하면서 바로 멤버를 세팅한다면 적절한 구조
nums, type(nums), len(nums)
```

```
nums = list() # 멤버를 추후에 동적으로 추가할 경우 적절
nums, type(nums), len(nums)
```

```
# 데이터 세팅하여 생성
nums = [1,3,5,7,9]
nums, type(nums), len(nums)
```

```
# 리스트의 멤버는 모두 같은 타입이다? => 거짓
nums = [1,3,5,7,9,'cat','dog']
nums, type(nums), len(nums)
```

```
nums = [1,3,5,7,9,'cat','dog', ['1','2']]
nums, type(nums), len(nums)
```

```
# 인덱싱 -> 변수[인덱스(정방향 or 역방향)] => 멤버가 나온다 -> 차원축소 -> 멤버(값)
# nums 안에 있는 '2'라는 멤버를 출력하시오
nums[-1][-1]
```

```
# 슬라이싱 -> 변수[ 시작인덱스:끝인덱스:step(기본값1) ] -> 차원유지 -> 리스트
# 1 <= x, step은 -1(역방향)
nums[ 1:-1 ]
```

```
# 원본 카피
nums[ : ], nums.copy()
```

```
# 원본값 수정
nums[ 1 ] = 'hi'
nums
```

```
# 값이 연속데이터타입(나열된 형태)라면 가능
nums[ 1:3 ] = 'hello'
```

```
nums
```

```
# 삭제
del nums[2]
nums
```

```
del nums[2:4]
nums
```

```
# 리스트 분해 -> 리스트를 변수로 받을 때, 변수가 나열되면 멤버 수와 동일해야 한다 
# 멤버들을 각각 변수에 직접 대입해서 사용할 수 있다 
# 멤버들의 정체성을 명확하게 부여할 수 있고, 인덱싱 사용 안 해도 됨
a, b = nums[3:5]
a, b
```

```
nums.append( '멤버추가' )
nums
```

```
a = list('ABC')
a
```

```
nums + a
```

```
nums.append(a)
```

```
nums
```

```
# 리스트에 또 다른 리스트를 이어붙이기 extend() : 원본조작 or + => 사본
b = nums[:-1]
b.extend( a )
b
```

```
# for문 같이 사용하면서 나오는 특징을 확인
# 자주 보이는 형태는 
# [ { 키:값,...}, {}, {} ] 
```

# 딕셔너리
- {}, 순서 X, 키와 값의 세트로 구성, 키는 중복되면 안 된다, 수정 ok, 삭제 ok, 추가 ok

```
dic = {}
dic, type(dic), len(dic)
```

```
dic = dict()
dic, type(dic), len(dic)
```

```
dic = {
    'name':'파이썬', # 1990
    'age':31
}
dic
```

```
# 인덱싱
dic[ 'name' ]
```

```
# 슬라이싱 X ← 순서가 없음
```

```
# 추가
# 키가 반드시 문자열일 필요는 없다! 함수는 안 됨
dic[1] = '17시'
len(dic), dic
```

```
print( dic )
```

# 튜플
- (), 순서가 존재(0,1,2,~ -1,-2,...), 값의 중복 ok, 삭제 x, 추가 x, 변경 x ← immutable
- 함수의 리턴값, 단순하게 값을 묶는다 

```
tu = ()
tu, type(tu), len(tu)
```

```
tu = tuple()
tu, type(tu), len(tu)
```

```
# 멤버가 1개면 반드시 ,를 붙인다
tu = (1,)
tu, type(tu), len(tu)
```

```
tu = (1,2,3,4)
```

```
# 인덱싱 
tu[0]
```

```
# 슬라이싱
tu[1:-1]
```

```
a = tu
a[0]
```

```
# 리스트나 튜플을 분해해서 변수로 직접 받고 싶다면, 
# 멤버 수대로 변수를 나열해야 한다. 이 중에서 사용하지 않는 변수는 통상 _로 표현
a , _ ,_a ,_b = tu
a # 1을 출력하고 싶다
```

```
# 튜플을 자주 쓰는 곳 
# 함수의 리턴값! <- 여러 개의 값을 리턴할 때!
```

# 집합
- {} . 키 X, 값만 존재, 값 중복 없음 -> 순서 없다
- 중복 제거

```
a = list('hellworld')
a
```

```
# 중복 제거
set(a)
# 합집합, 차집합, 교집합을 수행하고 => 리스트나 튜플형태로 빼서 다음 단계 진행
```

```
b = list(set(a))
b
```
