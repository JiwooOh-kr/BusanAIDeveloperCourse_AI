# if

- 조건에 따라 수행문을 달리 한다

```
# 현재 카페에 와서 커피를 구매했다
# 직원 曰 적립이나 할인카드 있나요? ... 2000원입니다. → 돈을 입력 받는다
money = input('적립이나 할인카드 있나요? ... 2000원입니다. ') 
```

```
money, type(money)
```

```
# 금액에 따라 처리하시오 → 조건을 구성해서 분기 처리, 케이스 등 생성 => if
'''
if 조건 :
  pass
elif 조건 :
  pass
elif 조건 :
  pass
else:
  pass
'''
```

```
money = int(money)
```

```
# 금액에 따라 처리하시오 => 상황은 총 3가지 존재 가능
# 가장 많이 존재할 수 있는 케이스 → 돈을 정확하게 내는 경우가 많다
money = 100
COFFEE_PRICE = 2000
if money == COFFEE_PRICE :
  print('결제 완료되었습니다.')
elif money > COFFEE_PRICE :
  print(f'결제 완료되었습니다. 잔돈은 { money-COFFEE_PRICE }원 입니다.')
else:
  print(f'결제금이 부족합니다. { COFFEE_PRICE - money }원 더 주시기 바랍니다.')
```

# while
- 반복문 : 동일한 작업을 반복할 때
- 0 ~ 무한대(언제 끝날지 모르겟다)
- break, if문이 항상 내부에 같이 움직일 확률이 높다

```
a = [1,2,3,4,5]
a
```

```
# a의 멤버가 존재할 때까지, 반복적으로 a의 멤버를 제거하시오
'''
while 조건:
  statements
'''
len(a)
```

```
while len(a) != 0 :
  del a[len(a)-1]
```

```
# len(a) => 5, 4, 3, 2, 1, 0:(조건식이 0이면 False) 
while len(a) :
  del a[len(a)-1]
```

```
# 리스트가 조건식에 들어가면, 멤버가 있으면 T, 멤버가 없으면 F
# 존재하면 xxx해라 라는 표현이 나오면 무조건 => 해당 변수를 조건식에 바로 기재하면 끝
a = [1,2,3,4,5]
while a :
  #del a[len(a)-1]
  print( a.pop() )
```

```
a = [1,2,3,4,5]
while a :  
  print( a.pop() )
else :
  print( '정상적으로 모든 순환 과정(반복 과정)을 수행했다' )
```

```
# 만약 a의 멤버수가 2개가 되면, 그만해라 
a = [1,2,3,4,5]
while a :  
  print( a.pop() )
  if len(a) == 2 :
    # 가장 가까운 반복문을 빠져나간다
    break
else :
  # 중도 하차했기 때문에 찍히지 않는다 ← 이런 부분의 체크가 필요하면 사용
  print( '정상적으로 모든 순환 과정(반복 과정)을 수행했다' )
```

# for
- 반복문
- 지정된 횟수를 반복한다 → 반복 횟수를 알고 있다
- 리스트, 딕녀서리, 튜플, 문자열등등과 자주 사용된다

```
a = [1,2,3,4,5,6,7,8,9,0]
a
```

```
# 방식 → for-each 방식만 가능
for item in a :
  print( item )
```

```
# 데이터 이중 구조
a = [ (1,2),(3,4),(5,6) ]
# a의 멤버 데이터가 튜플이므로 바로 분해해서 사용 가능
for k,v in a :
  print( v )
```

```
# 연속 수
for n in range(2,4) :
  print( n ) 
```

```
# 구구단을 구현하시오 (3 ~ 7단)
# 형식 '3 x 1 = 3', 2중 for문을 사용하시오
'''
3 x 1 = 3
3 x 2 = 6
...
7 x 9 = 63
# 출력되는 데이터가 각각 자리별로 주기를 가지고 반복된다
# 3,3,3,....3, 4,4,4,...4, ..... 7,7,7,...7
# 1,2,3,....9, 1,2,3,....9,....  1,2,3,....9
# 결과는 계산된 값이다
'''
```

```
# 3 → 4 → ... → 7
for x in range(3, 7+1) :      # 3~9
    for y in range(1, 9+1) :  # 1~9
      print( f'{x} x {y} = {x*y:2}' )  
# 결과를 리스트에 담아서 출력하시오
```

```
# 3 → 4 → ... → 7
for x in range(3, 7+1) :        # 3~9
    # 6단을 만나면, 7단으로 가라
    if x == 6 : continue
    for y in range(1, 9+1) :    # 1~9
      print( f'{x} x {y} = {x*y:2}' )  
# 결과를 리스트에 담아서 출력하시오
```

```
# 3 → 4 → ... → 7
for x in range(3, 7+1) :          # 3~9
    # 6단을 만나면, 7단으로 가라
    if x != 6 :
      for y in range(1, 9+1) :    # 1~9
        print( f'{x} x {y} = {x*y:2}' )  
# 결과를 리스트에 담아서 출력하시오
```

```
# 리스트 내포
# 6단만 빼고
[ f'{x} x {y} = {x*y:2}' for x in range(3, 7+1)  for y in range(1, 9+1)  ]
```
