# 게임 요구사항

- step 1
- 단, input()은 사용 가능(내장함수 사용가능)
- 게임이 진행되면 "게임 제목을 입력하시오" 프롬프트가 출력
- 사용자가 입력을 하고 엔터를 치면, 제목을 점검한다
- 만약 입력을 하지 않았다면
  - 정상적으로 입력하세요를 출력하고 다시 입력 대기
- 만약, 28자 이상 입력했다면
  - 게임 제목은 28자 이내입니다. 출력하고 다시 입력 대기
- 만약, 정상적으로 입력하고 28자 이내면(28자도 포함)
  - 다음 문구를 출력


```
# 게임 환경 변수
# 타이틀 최대 글자수
MAX_GAME_LEN = 28
IS_TEST_MODE = True
GAME_VERSION = 'v1.0.0'

while True :                        # 무한루프
  if IS_TEST_MODE :                 # 테스트 모드
    title = 'match number game'
  else :                            # 상용 모드
    title = input("게임 제목을 입력하시오")

  if not title :                    # 입력을 하지 않았다면
    print( '정상적으로 입력하세요' )
  elif len(title) > MAX_GAME_LEN :  # 28자 초과 입력했다면(28)
    print( f'입력하신 글자수는 {len(title)}입니다. {MAX_GAME_LEN}자이내로 입력하세요' )
  else :                            # 정상 입력
    break

# 프롬프트 출력
print( '여기까지 도착' )
prompt = f'''
==============================
={title:^28}=
={GAME_VERSION:^28}=
==============================
'''   # % MAX_GAME_LEN
print( prompt )
```

- step 2
- 게이머의 이름을 입력하세요? 프롬프트가 뜬다
  - 변수명 g_name
- 이름을 입력 후 엔터
- 이름을 입력하지 않고 엔터
  - "이름이 입력되지 않았습니다. 다시 입력하세요"
  - 입력 대기

```
# 변수를 초기화하고, 값이 없다는 것을 명시적으로 표현, 조건식 사용에도 유리
# 메모리 상에 존재한다
g_name = None
while not g_name :
  g_name = input('게이머의 이름을 입력하세요?')
  if not g_name :
    print( "이름이 입력되지 않았습니다. 다시 입력하세요" )
  else :
    break
print( g_name )
```

```
# 만약 while 안에서 다루는 변수가 다 변경되든지, 어떤 업무가 마무리되든지 하면 위의 방식은 안 된다. 실제로는 True를 쓰는 것도 좋지 않다.
# flag 변수를 활용
g_name = None
is_keep_nameInput = True
while is_keep_nameInput :
  g_name = input('게이머의 이름을 입력하세요?')
  if not g_name : 
    print( "이름이 입력되 않았습니다. 다시입력하세요" )
  else :
    is_keep_nameInput = False # 2중 반복문 내, depth 깊은 코드에서 전체 반복문을 빠져나갈 때 사용
    
print( g_name )
```

- 본 게임 시작
  - 사용자로부터 수치를 입력받는다
    - "0~99 사이의 값으로만 ai의 값을 예측하고 입력하세요" 입력대기
      - 아무것도 입력하지 않고 엔터
      - 숫자가 아닌 것을 입력하고 엔터
      - 숫자가 범위를 벗어나도록 입력하고 엔터
      - 정상적으로 입력하고 엔터
        - db에 로그 기록
    - AI가 0~99 사이의 숫자를 생성한다
    - 값체크
      - 입력값이 AI값보다 크다
      - 입력값이 AI값보다 작다 
      - 입력값이 AI값과 동일하다
        - 점수계산 100-(시도횟수*10)
        - "52 정답입니다. 총 시도회수는 4회이고, 획득한 점수는 60점입니다. 다시 하시겠습니까? yes or no로 입력하세요"
          - db에 로그 기록
            - 게이머 이름, 시도 횟수, 점수
              - 분석!(차후)
        - yes면 다시 게임 시작
        - no면 게임 종료
        - 이외의 상황인 경우는 정확하게 입력하세요
          - 입력 대기


```
targets = ['1', '1.1','a','가','!']
for n in targets :
  # 대상 문자열이 숫자가 될 수 있는 함수
  # n.isdecimal(), n.isdigit(), n.isnumeric()
  print( n, n.isalnum(), n.isdecimal(), n.isdigit(), n.isnumeric() )
```

```
while True :
  # 문자열을 입력 받으면 반드시 앞뒤 공백 제거 처리를 수행한다
  g_value = input( "0~99 사이의 값으로만 ai의 값을 예측하고 입력하세요" ).strip()
  # 점검
  if not g_value :                         # 아무것도 입력하지 않고, 공백은 조건식에서 False이니까 => False를 잡는 조건식
    print( '아무것도 입력하지 않았습니다. 다시' )
    pass
  elif not g_value.isnumeric() :           # 숫자가 아닌것을 입력하고 → False 
    print( '숫자가 아닙니다. 다시' )
    pass
  elif int(g_value)<0 or int(g_value)>99 : # 숫자가 범위를 벗어나게 입력
    print( '숫자가 범위를 벗어났습니다. 다시' )
    pass
  else :                                   # 정상
    break
    pass
```
