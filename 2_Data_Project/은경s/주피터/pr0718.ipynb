```python
from selenium import webdriver as wd
import time
import pandas as pd 
```

```python
chrome_options = wd.ChromeOptions()
chrome_options.add_argument("--incognito")

driver = wd.Chrome('./tool/chromedriver.exe', chrome_options=chrome_options)
driver.implicitly_wait(3)
```

# 팀 타격 메인

```python
pre_df2 = list()

for target_year in range (2015, 2021) :
    for target_month in range (3, 11) :
        target_site = f'http://www.kbreport.com/teams/main?teamId=&defense_no=&year_from={target_year}&year_to={target_year}&split01=month&split02_1={target_month}&split02_2={target_month}'
        driver.get(target_site)
        time.sleep(7)

        driver.find_elements_by_css_selector('#resultListDiv > table > tbody > tr')
        results = list()

        for td in driver.find_elements_by_css_selector('#resultListDiv > table > tbody > tr') :
            results.append(td.text.split())
                 
        if len(results) == 12 :
            for i in range (1, 11) :
                    pre_df2.append({
                            "연도" : target_year,
                            "월"   : target_month,
                            "순위" : results[i][0],
                            "팀명" : results[i][1],
                            '경기' : results[i][2], 
                            '승' : results[i][3], 
                            '무' : results[i][4], 
                            '패' : results[i][5], 
                            '승률' : results[i][6], 
                            '기대승률' : results[i][7],
                            'R/G' : results[i][8], 
                            '득점' : results[i][9], 
                            '홈런' : results[i][10], 
                            '도루' : results[i][11], 
                            '볼넷%' : results[i][12], 
                            '삼진%' : results[i][13], 
                            'BABIP' : results[i][14], 
                            '타율' : results[i][15], 
                            '출루율' : results[i][16], 
                            '장타율' : results[i][17], 
                            'OPS' : results[i][18],
                            'wOBA' : results[i][19], 
                            'WAR' : results[i][20]
                    }) 
    time.sleep(30)
```

```python
dfhm = pd.DataFrame(pre_df2)
dfhm
```

```python
dfhm.to_csv('kbreport_hit_main.csv', encoding='cp949')
```
