- 스크롤바를 하단으로 내려야 검색된 데이터가 더 나옴(ajax 사용해 데이터가 로드되므로 사용자 이벤트로 스크롤바를 이동시키는 기능 필요)

```python
# 1. 모듈 가져오기
from selenium import webdriver as wd
```

```python
# 2. 브라우저 띄우기
driver = wd.Chrome('./tool/chromedriver.exe')
```

```python
# 3. 접속
import urllib
keyword = "손흥민"
# 만약 한글 그냥 넣었을 때 안 되면 URL 인코딩 처리
# keyword = urllib.parse.quote(keyword)
target_site = f'https://www.youtube.com/results?search_query={keyword}&sp=CAISBAgCEAE%253D'
driver.get(target_site)
```

```python
# 4. 데이터를 모두 로드하기 위해 스크롤 처리 수행
# 단, 전제는 몇 번을 해야 다 로드되는지 모르기 때문에 1000번 수행(임의)
import time
# for n in range(1000) :
for n in range(10) :
    driver.execute_script('''
        window.scrollBy(0, 1000)
    ''')
    time.sleep(2)
```

```python
# 1. 검색된 컨텐츠 전체를 획득
videos = driver.find_elements_by_tag_name('ytd-video-renderer')
len(videos)
```

```python
for video in videos[:2] :
    target = video.find_element_by_id('video-title')
    # [▶]영상의 타이틀
    title = target.get_attribute('title')
    print(title)
    # [▶]썸네일
    thumbnail = video.find_element_by_id('img').get_attribute('src')
    print(thumbnail)
    # [▶]링크(실제 컨텐츠 url)
    href = target.get_attribute('href')
    print(href)
    # 작성자, 조회수, 작성시간
    pass
```

```python
contents = list()

for video in videos[:2] :
    target = video.find_element_by_id('video-title')
    contents.append({
        'title'     : target.get_attribute('title'),
        'thumbnail' : video.find_element_by_id('img').get_attribute('src'),
        'href'      : target.get_attribute('href')
    })
    pass
```

```python
# 메타 정보를 기준으로 상세 정보를 반복적으로 들어가면서 세부 데이터를 획득
# 웹상의 데이터를 수집해서 교육/연구/공부 쪽으로 사용하면 법적인 문제는 X but 상업적으로 이용하려면 법적 문제 확인해야함
for content in contents :
    # 해당 상세 페이지 접속
    driver.get(content['href'])
    # 대기
    # driver.implicitly_wait(50)
    time.sleep(5)
    # 리뷰를 더 불러오기(→ js로 스크롤 처리) : 생략
    # 리뷰수 획득
    # 리뷰 획득
```

```python
# 위의 영상은 최신순 정렬이라서 댓글이 얼마 없음, 다른 동영상으로 마저 진행
# 1. 접속(상세페이지)
url = 'https://www.youtube.com/watch?v=KXgrwNE9P1I'
driver.get(url)
time.sleep(2)
# 2. 스크롤 진행(리뷰를 로드시키고, 리뷰를 더 가져옴)
for n in range(10) :
    driver.execute_script('''
        window.scrollBy(0, 1000)
    ''')
    time.sleep(1)
```

```python
# 3. 리뷰 수집
for comment in driver.find_elements_by_tag_name('ytd-comment-thread-renderer') :
    print('-'*50)
    # 이 요소 밑에 있는 모든 텍스트를 가져와라
    print(comment.text)
    print('-'*50)
```
